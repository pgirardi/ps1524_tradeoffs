```python
# Import required libraries
import pandas as pd
```

```python
# Load the trp island report data
df = pd.read_csv("trp_island_gbks/trp_island_report.tsv", sep="\t")
df.head()
```

```python
# Clean up column names and sort by distance
df.columns = df.columns.str.strip()
df.sort_values(by=["span_bp"], ascending=False)
```

```python
# Remove leading/trailing whitespace from all string columns
str_cols = df.select_dtypes(include=['object']).columns
df[str_cols] = df[str_cols].apply(lambda x: x.str.strip())
```

```python
# Check value counts for trpG_label
df.trpG_label.value_counts()
```

```python
# Remove unwanted trpG_label entries
df = df[(df["trpG_label"] != "pyocyanine specific anthranilate synthase component II") & (df["trpG_label"] != "phnB")]
df.trpG_label.value_counts()
```

```python
# Check value counts for trpE_label
df.trpE_label.value_counts()
```

```python
# Remove unwanted trpG_label entries
df = df[(df["trpE_label"] == "trpE")]
df.trpE_label.value_counts()
```

```python
df.columns
```

```python
# Sort by distance and remove first 3 rows
df.sort_values(by=["span_bp"], ascending=False, inplace=True)
#df = df.iloc[3:, :]
```

```python
# Preview the DataFrame
df.head()
```

```python
# Search .gbk files for keywords and map to DataFrame
import os
from collections import defaultdict
gbk_dir = "trp_island_gbks"
keywords = ["phage", "capsid", "tail fiber", "baseplate"]
gbk_files = [f for f in os.listdir(gbk_dir) if f.endswith('.gbk')]
file_keyword_map = defaultdict(list)
for gbk_file in gbk_files:
    with open(os.path.join(gbk_dir, gbk_file), 'r') as f:
        content = f.read().lower()
        for kw in keywords:
            if kw in content:
                file_keyword_map[gbk_file].append(kw)
def find_matching_gbk(source_file):
    if not isinstance(source_file, str):
        return None
    prefix = source_file.replace('.gbk', '')
    for gbk_file in gbk_files:
        if gbk_file.startswith(prefix):
            return gbk_file
    return None
df['matched_gbk_file'] = df['source_file'].apply(find_matching_gbk)
def get_keywords_for_matched_file(row):
    fname = row['matched_gbk_file']
    if fname in file_keyword_map:
        return ', '.join(file_keyword_map[fname])
    return ''
df['gbk_keywords'] = df.apply(get_keywords_for_matched_file, axis=1)
df[['source_file', 'matched_gbk_file', 'gbk_keywords']].head()
```

```python
df.gbk_keywords.value_counts()
```

```python
# Show summary statistics for distance_bp
df.span_bp.describe()
```

```python
# Filter rows with distance_bp >= 10000
df = df[df["span_bp"] >= 10000]
df.shape
```

```python
# Save the cleaned DataFrame to a TSV file
df.to_csv("cleaned_trp_island_report.tsv", sep="\t", index=False)
```

```python
# Remove .gbk files not present in matched_gbk_file column
keep_files = set(df['matched_gbk_file'].dropna())
all_gbk_files = [f for f in os.listdir(gbk_dir) if f.endswith('.gbk')]
for gbk_file in all_gbk_files:
    if gbk_file not in keep_files:
        os.remove(os.path.join(gbk_dir, gbk_file))
        print(f"Removed: {gbk_file}")
print(f"Kept {len(keep_files)} .gbk files. Removed {len(all_gbk_files) - len(keep_files)} files.")
```
